## 다음 사항을 dataset 에 적용

1. 검증을 위해 훈련 세트의 데이터 일부 (예 : 10%)를 따로 설정하는 대신 테스트 세트를 검증 세트로 사용합니다. 이것은 훈련 할 데이터를 조금 더 제공합니다. 


2. 각 채널에서 평균을 빼고 표준 편차로 나누어 이미지 텐서를 정규화합니다. 결과적으로, 각 채널에 걸친 데이터의 평균은 0 이고 표준 편차는 1 입니다. 데이터를 정규화하면 한 채널의 값이 단순히 더 높거나 넓게 분산되어 (ex. 극단적으로 붉은 색의 사진) 다른 채널에 불균형적으로 영향을 주는 것을 방지할 수 있습니다.


3. 훈련 데이터 세트에서 이미지를로드하는 동안 무작위로 선택한 transformation 을 적용합니다. 특히 각 이미지를 4 픽셀씩 padding 한 다음 32 x 32 픽셀 크기로 임의의 cropping 을 수행한 다음 이미지를 50 % 확률로 가로로 뒤집습니다. transformation 은 특정 이미지가 로드될 때 마다 무작위로 동적으로 적용되므로 모델은 각 training epoch 마다 약간 다른 이미지를 보고 더 일반화 할 수 있습니다.

<img src="https://imgaug.readthedocs.io/en/latest/_images/cropandpad_percent.jpg">

```python
from torchvision.datasets import ImageFolder
import torchvision.transforms as tt
```

```python
# Data transforms (normalization & data augmentation)
stats = ((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))  # mean(r,g,b), std(r,g,b)
train_tfms = tt.Compose([tt.RandomCrop(32, padding=4, padding_mode='reflect'), 
                         tt.RandomHorizontalFlip(), 
                         tt.ToTensor(), 
                         tt.Normalize(*stats)])
valid_tfms = tt.Compose([tt.ToTensor(), tt.Normalize(*stats)])
```

```python
# PyTorch datasets
train_ds = ImageFolder(data_dir + '/train', train_tfms)
valid_ds = ImageFolder(data_dir + '/test', valid_tfms)
```

Next, we can create data loaders for retrieving images in batches.

```python
from torch.utils.data import DataLoader
```

```python
batch_size = 256
```

```python
# PyTorch data loaders
train_dl = DataLoader(train_ds, batch_size, shuffle=True, 
                      num_workers=8, pin_memory=True)
valid_dl = DataLoader(valid_ds, batch_size, shuffle=False, 
                      num_workers=8, pin_memory=True)
```

```python
len(train_dl) * 256, len(valid_dl) * 256
```

`num_workers` parameter 는 여러 CPU 코어를 활용하고 이미지를 병렬 로드하는 데 사용될 수 있습니다. 'pin_memory' parameter 는 각 데이터 배치를 로드하기 위해 동일한 메모리 부분(RAM)을 사용하여 반복되는 메모리 할당 및 할당 해제를 피합니다. 모든 이미지의 크기가 같기 때문에(32px x 32px) 이 작업을 수행 할 수 있습니다. 파이프 라인의 이러한 작은 변경으로 인해 성능이 크게 향상되었습니다.

```python
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
```

```python
def show_batch(dl):
    for images, labels in dl:
        fig, ax = plt.subplots(figsize=(16, 16))
        ax.set_xticks([]); ax.set_yticks([])
        ax.imshow(make_grid(images[:100], 10).permute(1, 2, 0))
        break
```

```python
show_batch(train_dl)
```

정규화로 인해 색상이 다르게 보입니다. 정규화는 inference 시에도 적용됩니다. 자세히 보면 일부 이미지에서 자르기 및 reflection padding 을 볼 수 있습니다. 수평 플립은 육안 검사로는 감지하기가 약간 어렵습니다.

## Model with Residual Blocks and Batch Normalization

We'll use a significatly larger model this time, called the WideResNet22, which has 22 convolutional layers. However, one of the key changes to our model is the addition of the **resudial block**, which adds the original input back to the output feature map obtained by passing the input through one or more convolutional layers.

이번에는 22 개의 컨볼 루션 레이어가있는 WideResNet22 라는 상당히 큰 모델을 사용합니다. 그러나 모델의 주요 변경 사항 중 하나는 **resudial 블록** 을 추가하는 것입니다.이 블록은 입력을 하나 이상의 convolutional layer 를 통해 전달하여 얻은 출력 feature map 에 원래 입력을 다시 추가합니다. (skip-connection)

<img src="https://miro.medium.com/max/1140/1*D0F3UitQ2l5Q0Ak-tjEdJg.png">

F(x) - residual

아래는 매우 간단한 Residual block 입니다.

```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class SimpleResidualBlock(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=3, kernel_size=3, stride=1, padding=1)
        self.relu1 = nn.ReLU()
        self.conv2 = nn.Conv2d(in_channels=3, out_channels=3, kernel_size=3, stride=1, padding=1)
        self.relu2 = nn.ReLU()
        
    def forward(self, x):
        out = self.conv1(x)
        out = self.relu1(out)
        out = self.conv2(out)
        return self.relu2(out + x)
```

```python
simple_resnet = SimpleResidualBlock()

for images, labels in train_dl:
    out = simple_resnet(images)
    print(out.shape)
    break
```

이 작은 변화는 모델의 성능을 크게 향상시킵니다. residual blocks 이 어떻게 작동하는지 참조: https://towardsdatascience.com/residual-blocks-building-blocks-of-resnet-fd90ca15d6ec .

또한 각 컨볼루션 레이어 뒤에 batch normalization 레이어를 추가합니다. 배치 정규화에 대한 좋은 설명이 있습니다: https://towardsdatascience.com/batch-normalization-and-dropout-in-neural-networks-explained-with-pytorch-47d7a8459bcd


다음과 같은 구조를 갖는 WideResNet22 네트워크를 만들기위한 일련의 도우미 함수를 정의합니다.

- Group1 의 1x1 convolution 은 channel 을 늘리기 위해 사용 (16 -> 96)
- Group2, 3 에서는 pooling 대신 stride 2 적용 (image size 를 반으로 축소)  
- Adaptive Average Pooling 은 output size 를 미리 정해 놓고 pooling 

<img src="https://miro.medium.com/max/2500/1*gOyiFYGHMAEDlyoAIhkWkA.png">


ni - # of input channels  
nf - # of output channels

```python
def conv_2d(ni, nf, stride=1, ks=3):
    return nn.Conv2d(in_channels=ni, out_channels=nf, 
                     kernel_size=ks, stride=stride, 
                     padding=ks//2, bias=False)

def bn_relu_conv(ni, nf):
    return nn.Sequential(nn.BatchNorm2d(ni), 
                         nn.ReLU(inplace=True), 
                         conv_2d(ni, nf))

class ResidualBlock(nn.Module):
    def __init__(self, ni, nf, stride=1):
        super().__init__()
        self.bn = nn.BatchNorm2d(ni)
        self.conv1 = conv_2d(ni, nf, stride)
        self.conv2 = bn_relu_conv(nf, nf)
        self.shortcut = lambda x: x
        if ni != nf:
            self.shortcut = conv_2d(ni, nf, stride, 1)
    
    def forward(self, x):
        x = F.relu(self.bn(x), inplace=True)
        r = self.shortcut(x)
        x = self.conv1(x)
        x = self.conv2(x) * 0.2
        return x.add_(r)
```

```python
def make_group(N, ni, nf, stride):
    start = ResidualBlock(ni, nf, stride)
    rest = [ResidualBlock(nf, nf) for j in range(1, N)]
    return [start] + rest

class Flatten(nn.Module):
    def __init__(self): super().__init__()
    def forward(self, x): return x.view(x.size(0), -1)

class WideResNet(nn.Module):
    def __init__(self, n_groups, N, n_classes, k=1, n_start=16):
        super().__init__()      
        # Increase channels to n_start using conv layer
        layers = [conv_2d(3, n_start)]
        n_channels = [n_start]
        
        # Add groups of BasicBlock(increase channels & downsample)
        for i in range(n_groups):
            n_channels.append(n_start*(2**i)*k)
            stride = 2 if i>0 else 1
            layers += make_group(N, n_channels[i], 
                                 n_channels[i+1], stride)
        
        # Pool, flatten & add linear layer for classification
        layers += [nn.BatchNorm2d(n_channels[3]), 
                   nn.ReLU(inplace=True), 
                   nn.AdaptiveAvgPool2d(1), 
                   Flatten(), 
                   nn.Linear(n_channels[3], n_classes)]
        
        self.features = nn.Sequential(*layers)
        
    def forward(self, x): return self.features(x)
    
def wrn_22(): 
    return WideResNet(n_groups=3, N=3, n_classes=10, k=6)
```

```python
model = wrn_22()
```

```python
for images, labels in train_dl:
    print('images.shape [batch, channel, height, width]:', images.shape)
    out = model(images)
    print('out.shape: [batch, output]', out.shape)
    break
```

## Training the model

PyTorch 위에 구축 된 FastAI 라이브러리는 learning rate scheduling 기능이 있는 강력하고 사용자 정의 가능한 훈련 루프를 포함하여 많은 유틸리티를 제공합니다.

Learning rate scheduling 은 모델이 훈련되는 동안 학습 속도를 동적으로 변경하는 프로세스를 말합니다. 훈련 중에 학습 속도를 변경하기위한 많은 전략이 있으며, 우리가 사용할 전략을 1-cycle policy 라고 합니다. 낮은 학습 속도로 시작한 다음 훈련의 약 절반 동안 점차적으로 높은 값으로 증가시킨 다음 천천히 원래 값으로 가져옵니다. 결국, 우리는 매우 낮은 학습 속도로 몇 번의 반복 훈련을 합니다.

<img src="https://sgugger.github.io/images/art5_lr_schedule.png">

1-cycle policy 설명: https://sgugger.github.io/the-1cycle-policy.html


```python
from fastai.basic_data import DataBunch
from fastai.train import Learner
from fastai.metrics import accuracy
```

먼저 이전에 만든 데이터 로더를 사용하여 `DataBunch` 개체를 만든 다음 data bunch, 모델 및 손실 함수를 사용하여 `Learner` 개체를 만듭니다.

```python
data = DataBunch.create(train_dl, valid_dl, bs=batch_size, path=data_dir)
#data = DataBunch(train_dl, valid_dl)
learner = Learner(data, model, loss_func=F.cross_entropy, metrics=[accuracy])
learner.clip = 0.1
```

`learner.clip = 0.1` 은 그래디언트 클리핑을 수행하는 데 사용됩니다. 즉, 그래디언트 값을`[-0.1,0.1]`범위로 제한하여 큰 값으로 인한 매개 변수 (가중치 및 바이어스)의 바람직하지 않은 변경을 방지합니다. 이렇게 하여 특정 feature 의 gradient 가 전체를 지배하지 못하도록 합니다.

FastAI는 또한 다양한 학습 속도를 시도하는 learning rate finder 를 제공하며, 손실 대 learning rate 의 그래프를 보면서 좋은 학습 속도를 선택할 수 있습니다.

```python
learner.lr_find()
```

Based on the above graph, it seems like a learning rate of `5e-3` leads to the fastest reduction in loss.

To train the model using the 1-cycle policy, we use the `.fit_one_cycle` method, and pass in the no. of epochs and the maximum learning rate. We also use [weight decay](https://towardsdatascience.com/this-thing-called-weight-decay-a7cd4bcfccab), which is yet another regularization technique which prevents the weights from becoming too large by adding an additional term to the loss function.

```python
learner.fit_one_cycle(9, 5e-3, wd=1e-4)
```

As you can see, the model quickly reaches an accuracy of over 90% in just 9 epochs. It is quite a feat, considering that our simple CNN model struggled to reach 60% accuracy even after training for a long time.

In fact, if we continue to train this model for longer, it will soon reach an accuracy of around 95%, which is quite close to the state of the art. The model and techniques covered in this tutorial were actually used by the FastAI team for their submission to the Stanform DawnBench competition.
