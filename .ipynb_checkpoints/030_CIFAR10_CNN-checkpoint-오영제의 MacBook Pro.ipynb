```python
import numpy as np
import torch
from torch import nn, optim
import torchvision
from torchvision import transforms
import torch.nn.functional as F
import matplotlib.pyplot as plt
from torchvision.utils import make_grid
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

## Data Download 및 Data Loader 를 이용하여 Train, Validation data 준비

```python
transform = transforms.Compose([
                transforms.ToTensor(),
                transforms.Normalize((0.5, 0.5, 0.5),(0.5, 0.5, 0.5))   # (mean, std)
                                    ])

train_dataset = torchvision.datasets.CIFAR10(root='./data', 
                train=True, download=True, transform=transform)                
validation_dataset  = torchvision.datasets.CIFAR10(root='./data', 
                train=False, download=True, transform=transform)
```

```python
EPOCHS = 50
BATCH_SIZE = 512
lr = 1e-3

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
validation_loader = torch.utils.data.DataLoader(validation_dataset, batch_size=BATCH_SIZE * 2, shuffle=False)
```

```python
print(train_dataset)
print(train_dataset[0][0].shape)
print(train_dataset[0][1])
print()
print(validation_dataset)
print(validation_dataset[0][0].shape)
print(validation_dataset[0][1])
```

```python
cifa10_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
                  'dog', 'frog', 'horse', 'ship', 'truck']
```

### 일부 Data 시각화

```python
fig, ax = plt.subplots(figsize=(10, 10))

for images, labels in train_loader:
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(make_grid(images[:100], 10).permute(1, 2, 0))
    break
```

```python
dataiter = iter(train_loader)
images, labels = dataiter.next()

fig, axes = plt.subplots(2, 8, figsize=(15, 4))
axes = axes.ravel()
for i in range(16):
    img = np.transpose(images[i], (1, 2, 0)) / 2 + 0.5    # unnormalize
    axes[i].imshow(img)
    axes[i].set_xticks([])
    axes[i].set_yticks([])
    axes[i].set_title(cifa10_classes[labels[i]])

plt.show()
```

### Model Build

```python
def conv_output(input_size, filter_size, padding_size, stride):
    return np.floor((input_size - filter_size + 2*padding_size) / stride + 1)

def maxpool(input_size, kernel_size, stride):
    return np.floor((input_size - kernel_size) / stride + 1)

print(conv_output(32, 3, 1, 1))
print(maxpool(32, 2, 2))
print(conv_output(16, 3, 1, 1))
print(maxpool(16, 2, 2))
print(conv_output(8, 3, 1, 1))
print(maxpool(8, 2, 2))
```

### nn.Module 이용

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
        self.pool1  = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.pool2  = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool3  = nn.MaxPool2d(2, 2)
        self.fc1      = nn.Linear(64 * 4 * 4, 120)
        self.fc2      = nn.Linear(120, 84)
        self.fc3      = nn.Linear(84, 10)
        
    def forward(self, x):
        x = self.pool1(torch.relu(self.conv1(x)))
        x = self.pool2(torch.relu(self.conv2(x)))
        x = self.pool3(torch.relu(self.conv3(x)))
        x = x.view(-1, 64 * 4 * 4)
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

### nn.Sequential 이용

```python
sequential = nn.Sequential(
    nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2, 2), # output: bs x 16 x 16 x 16
    
    nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2, 2), # output: bs x 32 x 8 x 8

    nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
    nn.ReLU(),
    nn.MaxPool2d(2, 2), # output: bs x 64 x 4 x 4
    
    nn.Flatten(),
    
    nn.Linear(1024, 120),  # output: bs * 120
    nn.Linear(120, 84),
    nn.Linear(84, 10) # output: bs x 10 
)
```

```python
#model = Net()
model = sequential
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
criterion = criterion.to(device)
```

```python
optimizer = optim.Adam(model.parameters(), lr=lr)
```

```python
model.train()
```

```python
import time
st = time.time()
LOSS = []
VAL_LOSS = []
ACCURACY = []

for epoch in range(EPOCHS):
    cost = 0
    model.train()
    for x, y in train_loader:
        x, y = x.to(device), y.to(device)    
        z = model(x)
        loss = criterion(z, y)
        
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
        cost += loss.data

    LOSS.append(cost)
        
    correct = 0
    val_cost = 0
    with torch.no_grad():
        for x, y in validation_loader:
            x, y = x.to(device), y.to(device)

            z = model(x)
            
            val_loss = criterion(z, y)
            val_cost += val_loss
            
            _, label = torch.max(z, 1)
            correct += (label == y).sum().item()

    VAL_LOSS.append(val_cost)
        
    accuracy = 100 * (correct / len(validation_dataset))
    ACCURACY.append(accuracy)
    
    print(f'epoch {epoch} ==> train loss: {cost}, ' + 
                             f'val loss: {val_cost}, ' +
                             f'valididation accuracy: {accuracy}')

et = time.time()
print(et - st)
```

### model evaluation

```python
model.eval()
```

### Model Evaluation

```python
correct = 0

for x, y in validation_loader:
    x, y = x.to(device), y.to(device)
    z = model(x)
    _, label = torch.max(z, 1)
    correct += (label == y).sum().item()
    
print(f'evaluation accuracy: {correct / len(validation_dataset) * 100: .2f}')
```

```python
plt.figure(figsize=(8, 4))

fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.plot(LOSS, color=color)
color = 'tab:green'
ax1.plot(VAL_LOSS, color=color)
ax1.set_xlabel('epoch', color=color)
ax1.set_ylabel('total loss', color=color)

color = 'tab:blue'
ax2 = ax1.twinx()
ax2.plot(ACCURACY, color=color)
ax2.set_ylabel('accuracy', color=color);
```

### model 이 어떤 image 들을 잘 맞추고 혹은 틀렸는지 시각화

```python
%matplotlib inline

plt.figure(figsize=(7, 7))
count = 0

for x, y in torch.utils.data.DataLoader(dataset=validation_dataset, batch_size=1):
    if count > 8:
        break
    x = x.to(device)
    z = model(x)
    _, yhat = torch.max(z, 1)
    yhat = yhat.to("cpu")
    if yhat == y:
        plt.subplot(3, 3, count+1)
        x = make_grid(x.to("cpu"))
        x = np.transpose(x, (1, 2, 0)) / 2 + 0.5
        plt.imshow(x, cmap='gray', interpolation='none')  
        plt.title("{}, {}".format(cifa10_classes[y.item()], cifa10_classes[yhat.item()]))
        plt.tight_layout()
        plt.xticks([])
        plt.yticks([])
        count += 1

plt.figure(figsize=(7, 7))
count = 0

for x, y in torch.utils.data.DataLoader(dataset=validation_dataset, batch_size=1):
    if count > 8:
        break
    x = x.to(device)
    z = model(x)
    _, yhat = torch.max(z, 1)
    yhat = yhat.to("cpu")
    if yhat != y:
        plt.subplot(3, 3, count+1)
        x = make_grid(x.to("cpu"))
        x = np.transpose(x, (1, 2, 0)) / 2 + 0.5
        plt.imshow(x, cmap='gray', interpolation='none')  
        plt.title("{}, {}".format(cifa10_classes[y.item()], cifa10_classes[yhat.item()]))
        plt.tight_layout()
        plt.xticks([])
        plt.yticks([])
        count += 1
```

### 각 이미지별 정확도

```python
class_correct = list(0. for _ in range(10))
class_total   = list(0. for _ in range(10))
y_pred = []
y_true = []

for x, y in validation_loader:
    x, y = x.to(device), y.to(device)
    z = model(x)
    _, yhat = torch.max(z, 1)

    y_pred.extend(yhat.to('cpu').numpy())
    y_true.extend(y.to('cpu').numpy())

    c = yhat == y
    for i in range(len(x)):
        label = yhat[i]
        class_correct[label] += c[i]
        class_total[label] += 1

for i in range(10):
    print(f"{cifa10_classes[i]} 의 accuracy : {100 * class_correct[i]/class_total[i]:.2f} %")
```

### confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns

cm = confusion_matrix(y_true, y_pred)
cm
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(cm, annot=True, fmt='d')

plt.xticks(np.arange(10), cifa10_classes, rotation=45, fontsize=12)
plt.yticks(np.arange(10), cifa10_classes, rotation=45, fontsize=12)
plt.xlabel("true class")
plt.ylabel("predicted class")
plt.title('Confusion Matrix')
print('Test Accuracy :', accuracy_score(y_true, y_pred))
```
